#### 4.  Read the dataset ‘dementia.csv’, where the notations are shown as follows.
#### age: age range
#### total: population by age range
#### dementia: population with dementia


(a) Please make a stacked bar chart with the x-axis as the age range and the y-axis as the
 population.
 Hint: Assign colors to bars using "color='b'" for blue and "color='c'" for cyan.

```python
dementia = pd.read_csv("./data/dementia.csv")
```

```python
dementia.head()
```

```python
plt.bar(height=dementia["dementia"], x=dementia["age"], color="c", label="dementia")
plt.bar(height=dementia["total"], x=dementia["age"], bottom=dementia["dementia"], color="b", label="total")
plt.xlabel("Age")
plt.ylabel("Population")
plt.title('Dementia Demographics')
plt.legend()
plt.show()
```

##### b) add the information of proportion by line chart on the stack bar chart.
 Hint: <br>
 (1) You may use “Axes.twinx()” to create a twin axis sharing the x-axis. <br>
 (2)You may use “Axes.set_ylim(0, 100)” to set the range of proportion from 0 to 100.

```python
plt.bar(height=dementia["dementia"], x=dementia["age"], color="c")
plt.bar(height=dementia["total"], x=dementia["age"], bottom=dementia["dementia"], color="b")
plt.xlabel("Age")
plt.ylabel("Population")
plt.title('Dementia Demographics')
plt.legend()


# Create a twin axis for the line chart
ax2 = plt.twinx()

# Calculate proportion 
proportion = [100 * dem / total if total != 0 else 0 for dem, total in zip(dementia["dementia"], dementia["total"])]
ax2.plot(dementia["age"], proportion, color="r", marker="o", linestyle="-", label="Proportion")

# Set y-axis limit for proportion
ax2.set_ylim(0, 100)

# Add legend for the line chart
ax2.legend(loc="upper right")
ax2.set_ylabel("Proportion (%)", color="red")
plt.show()
```

## Sharing Axis

```python
 x = np.linspace(0, 2*np.pi, 400)
 y = np.sin(x**2)
```

```python
 f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
 ax1.plot(x, y)
 ax2.scatter(x, y)
 ax1.set_title('Sharing Y axis')
 plt.show()
```

```python
 fig, (ax3, ax4) = plt.subplots(1, 2,subplot_kw=dict(projection="polar"))
 ax3.plot(x, y)
 ax4.scatter(x, y)
 plt.show()
```
